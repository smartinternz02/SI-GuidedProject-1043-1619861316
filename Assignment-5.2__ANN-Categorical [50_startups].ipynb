```python
df.Administration=df.Administration.fillna(df.Administration.mean())
```

```python
df["Marketing Spend"]=df["Administration"].fillna(df["Marketing Spend"].mean())
```

```python
x=df.iloc[:,[0,1,2,4]].values
y=df.iloc[:,3:4].values
```

```python
x.shape
```

```python
y.shape
```

```python
from sklearn.preprocessing import OneHotEncoder
from sklearn.compose import ColumnTransformer
one=OneHotEncoder()
z=one.fit_transform(y[:,0:1]).toarray()
y=np.delete(y,0,axis=1)
y=np.concatenate((z,y),axis=1)
```

```python
y.shape
```

```python
y
```

```python
from sklearn.model_selection import train_test_split
x_train,x_test,y_train,y_test = train_test_split(x , y , test_size=0.2 , random_state=100)
```

```python
x_train=np.asarray(x_train).astype('float32')
x_test=np.asarray(x_test).astype('float32')
y_train=np.asarray(y_train).astype('float32')
y_test=np.asarray(y_test).astype('float32')
```

```python
from sklearn.preprocessing import StandardScaler
sc=StandardScaler()
x_train=sc.fit_transform(x_train)
x_test=sc.fit_transform(x_test)
```

```python
from tensorflow.keras.models import Sequential   
from tensorflow.keras.layers import Dense
```

```python
b=Sequential()
```

```python
b.add(Dense(4,kernel_initializer= "random_uniform",activation="relu"))
```

```python
b.add(Dense(8,kernel_initializer= "random_uniform",activation="relu"))
```

```python
b.add(Dense(3,kernel_initializer= "random_uniform",activation="softmax"))
```

```python
b.compile("adam","categorical_crossentropy",metrics=["accuracy"])
```

```python
b.fit(x_train,y_train,batch_size=32,epochs=1000)
```

```python
ypred=b.predict(x_test)
ypred
```
