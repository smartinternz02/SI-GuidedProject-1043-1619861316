```python
from sklearn.preprocessing import StandardScaler
sc=StandardScaler()
x_train=sc.fit_transform(x_train)
x_test=sc.fit_transform(x_test)
```

```python
from tensorflow.keras.models import Sequential   
from tensorflow.keras.layers import Dense
```

```python
a=Sequential()
```

```python
a.add(Dense(8,kernel_initializer= "random_uniform",activation="relu"))
```

```python
a.add(Dense(2,kernel_initializer= "random_uniform",activation="softmax"))
```

```python
a.compile("adam","categorical_crossentropy",metrics=["accuracy"])
```

```python
a.fit(x_train,y_train,batch_size=32,epochs=1000)
```
